```python
lamb = 1
QUBO_all = np.load(f"./QUBO_L{lamb}.npz")
asset = 6 # 3, 6, 9
E = 10 # 0, 1, ..., 34

QU = QUBO_all[f"A{asset}_E{E}"]
print(QU.shape)

qb = 2*asset
ll = np.zeros((qb, 1<<qb), dtype=np.float32)
a_0 = np.zeros(1<<qb, dtype=np.float32)
a_1 = np.ones(1<<qb, dtype=np.float32)
idxx = np.arange(1<<qb, dtype=np.int32)
for i in range(qb):
    ll[i] = np.where(idxx%(1<<(qb-i))<(1<<(qb-i-1)), a_0,  a_1)
l = ll.T.copy()
ss = l @ QU
ss = lamb - (ss.reshape(-1, 1, qb) @ l.reshape(-1, qb, 1)).reshape(-1)

print(ss.shape)
print(ss.min(), ss.max(), ss.mean())
```
